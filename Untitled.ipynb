```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
canvas = np.zeros(shape = (512,512,3), dtype = "int")
```

```python
cv2.rectangle(canvas, pt1 = (5,5), pt2 = (100,100), color = (0,255,0), thickness = 2)
plt.imshow(canvas)
```

```python
cv2.circle(canvas, center = (350,100), radius = 50, color = (0,0,255), thickness = 5)
plt.imshow(canvas)
```

```python
cv2.line(canvas, pt1 = (10,500), pt2 = (500,10), thickness= 5, color = (255,0,0))
plt.imshow(canvas)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(canvas, text = "hello", org = (100,500), color = (0,0,255), thickness = 2, fontFace = font, fontScale = 4)
plt.imshow(canvas)
```

```python
image = cv2.imread("DATA/rainbow.jpg")
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV", hsv)
cv2.imshow("Hue Channel", hsv[:,:,0])
cv2.imshow("Saturation Channel", hsv[:,:,1])
cv2.imshow("Value Channel", hsv[:,:,2])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image[:,:,0] = 0
image[:,:,1] = 0
cv2.imshow("red", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image = cv2.imread("DATA/rainbow.jpg")
color=  ("b", "g", "r")
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
plt.show()
```
